```python
#column renames
df.rename(columns={
    'full_name': 'Object_Name',
    'a': 'Semimajor_Axis_AU',             #Semi-major axis (average orbital distance)
    'e': 'Eccentricity',                       #Orbital eccentricity
    'i': 'Inclination_deg',              #Orbital inclination in degrees
    'om': 'Longitude_Ascending_Node_deg', #Longitude of the ascending node in degrees
    'w': 'Argument_of_Perihelion_deg', #Argument of perihelion in degrees
    'q': 'Perihelion_Distance_AU',        #Closest approach to the Sun
    'ad': 'Aphelion_Distance_AU',          #Farthest distance from the Sun
    'per_y': 'Orbital_Period_years',  #Time to complete one orbit around the Sun
    'data_arc': 'Data_Arc_Span_days', #Observation time span in days
    'condition_code': 'Condition_Code',        #Code indicating uncertainty of the orbit
    'n_obs_used': 'Observations_Used_Total', #Total number of observations
    'n_del_obs_used': 'Observations_Used_Delay',  #Number of delay (radar) observations
    'n_dop_obs_used': 'Observations_Used_Doppler',#Number of Doppler observations
    'H': 'Absolute_Magnitude'                  #Apparent brightness at standard distance
}, inplace=True)
df.head()
```

```python
#column dropping
droppedColumns = ['Observations_Used_Delay', 'Observations_Used_Doppler', 'Condition_Code']
df.drop(labels=droppedColumns, axis=1, inplace=True)
df.head()
```

```python
#renaming columns
dtypes = {
    'Object_Name': 'object',
    'Semimajor_Axis_AU': 'float64',
    'Eccentricity': 'float64',
    'Inclination_deg': 'float64',
    'Longitude_Ascending_Node_deg': 'float64',
    'Argument_of_Perihelion_deg': 'float64',
    'Perihelion_Distance_AU': 'float64',
    'Aphelion_Distance_AU': 'float64',
    'Orbital_Period_years': 'float64',
    'Data_Arc_Span_days': 'Int64',
    'Observations_Used_Total': 'Int64',
    'Absolute_Magnitude': 'float64',
}

for col, dtype in dtypes.items():
    df[col] = df[col].astype(dtype)

print(df.dtypes)
```

```python
#getting rid of nans in Data_Arc_Span_days
print("NaN count before fill:", df['Data_Arc_Span_days'].isna().sum())
df['Data_Arc_Span_days'] = df['Data_Arc_Span_days'].astype(float)
mean_value = df['Data_Arc_Span_days'].mean()
df['Data_Arc_Span_days'] = df['Data_Arc_Span_days'].fillna(mean_value)
df['Data_Arc_Span_days'] = df['Data_Arc_Span_days'].round().astype(int)
print("NaN count after fill:", df['Data_Arc_Span_days'].isna().sum())
```

```python
#getting rid of nans in Absolute_Magnitude
print("NaN count before fill:", df['Absolute_Magnitude'].isna().sum())
df['Absolute_Magnitude'] = df['Absolute_Magnitude'].astype(float)
mean_value = df['Absolute_Magnitude'].mean()
df['Absolute_Magnitude'] = df['Absolute_Magnitude'].fillna(mean_value)
print("NaN count after fill:", df['Absolute_Magnitude'].isna().sum())
```

```python
#getting rid of nans in Aphelion_Distance_AU
print("NaN count before fill:", df['Aphelion_Distance_AU'].isna().sum())
df['Aphelion_Distance_AU'] = df['Aphelion_Distance_AU'].astype(float)
mean_value = df['Aphelion_Distance_AU'].mean()
df['Aphelion_Distance_AU'] = df['Aphelion_Distance_AU'].fillna(mean_value)
print("NaN count after fill:", df['Aphelion_Distance_AU'].isna().sum())
```

```python
#getting rid of nans in Orbital_Period_years            
print("NaN count before fill:", df['Orbital_Period_years'].isna().sum())
df['Orbital_Period_years'] = df['Orbital_Period_years'].astype(float)
mean_value = df['Orbital_Period_years'].mean()
df['Orbital_Period_years'] = df['Orbital_Period_years'].fillna(mean_value)
print("NaN count after fill:", df['Orbital_Period_years'].isna().sum())
```

```python
#plotting example asteroid orbits
import numpy as np
import matplotlib.pyplot as plt

def plot_orbit(a, e, label=None):
    theta = np.linspace(0, 2 * np.pi, 500) #angle in radians
    r = a * (1 - e**2) / (1 + e * np.cos(theta)) #Polar ellipse equation, in this equation, a is semimajor axis, e is eccentricity, r is distance from the focus (sun), at angle theta
    x = r * np.cos(theta) #Converting to cartesian (regular coordinate form)
    y = r * np.sin(theta) #these represent points along the orbit, which are plotted and joined up
    plt.plot(x, y, label=f'a={a}, e={e}' if label is None else label) #plotting the orbit

#setting up the graph
plt.figure(figsize=(8, 8))
plt.title("Shapes of Orbits with Different Semimajor Axis and Eccentricity")
plt.xlabel("x (AU)")
plt.ylabel("y (AU)")
plt.axhline(0, color='grey', linestyle='--')
plt.axvline(0, color='grey', linestyle='--')

#plotting the orbits with varying semimajor axis and eccentricity 
plot_orbit(1, 0.0, label="Circle")
plot_orbit(1, 0.3)
plot_orbit(1, 0.6)
plot_orbit(1, 0.9)
plot_orbit(1, 0.99) #were keeping a constant to see how eccentricity impacts orbits

plt.legend()
plt.grid(True)
plt.axis('equal')  # Equal scaling for x and y to show true orbit shapes
plt.show()
```

```python
#plotting mean asteroid values as orbits
import numpy as np
import matplotlib.pyplot as plt

# Define the semimajor axis and eccentricity for each category
categories = {
    'Bright': {'a': 55.747796, 'e': 0.202163},
    'Medium': {'a': 3.738361, 'e': 0.128370},
    'Dim': {'a': 2.675654, 'e': 0.160225}
}

# Define colors for each category
colors = {'Bright': 'orange', 'Medium': 'green', 'Dim': 'red'}

# Plot settings
plt.figure(figsize=(10, 8))

# Plot each orbit
for category, params in categories.items():
    a = params['a']
    e = params['e']
    theta = np.linspace(0, 2 * np.pi, 500)
    r = a * (1 - e ** 2) / (1 + e * np.cos(theta))
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    
    # Plot the orbit with unique color and label
    plt.plot(x, y, color=colors[category], label=f"{category} (a={a:.2f}, e={e:.5f})", linewidth=2)

# Highlight the Sun at the center
plt.plot(0, 0, 'yo', markersize=8, label='Sun')

# Customize the plot
plt.title("Orbits of the Average Asteroid by Brightness Category")
plt.xlabel("x (AU)")
plt.ylabel("y (AU)")
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.axis('equal')
plt.show()
```
